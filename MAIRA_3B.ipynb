Exibindo localização de academias de artes marciais em Curitiba. Dados filtrados do arquivo Base de Alvarás, disponível em https://dadosabertos.curitiba.pr.gov.br/

```python
academias_artes_marciais = '/content/drive/My Drive/GA155/academias_geocode.geojson'

df = gpd.read_file(academias_artes_marciais)

#df

med_lat = df.geometry.y.mean()
med_lon = df.geometry.x.mean()

map = folium.Map(location=[med_lat, med_lon],
                 tiles='OpenStreetMap',
                 zoom_start=12)

for idx, row in df.iterrows():
    if row['geometry'] is not None:
        folium.Marker(
            location=[row['geometry'].y, row['geometry'].x],
            tooltip=row['NOME_FANTASIA'],
            icon=folium.Icon(color='orange', icon='certificate')
        ).add_to(map)

map
```
